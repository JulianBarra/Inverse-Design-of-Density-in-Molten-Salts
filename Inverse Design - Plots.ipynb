```python
#Low Density
#presen_HD = [-355.0044095384335, meanen_MNS_LD, meanen_DEF_LD, -353.58650130550694, -352.6600835738668]#]
#Volsen_HD = [2.287715184, 2.5055915040038176, 2.6308709812104476, 3.025501629, 3.4201]
#pres_HD = [mean_MNS_LD, mean_DEF_LD, ]#]
#
pres_LD = [19.201555, 10.541676519271695, 6.483100245720712, 1.5297822510403563, -0.6640897544453852]#]
Vols_LD = [2.287715184, 2.5055915040038176, 2.6308709812104476, 3.025501629, 3.4201]

#Mid Density
#presen_HD = [mean_MNS_MD, mean_DEF_MD, -454.87963098519845]
#Volsen_HD = [1.520281143, 1.59629509, 1.83573934]
pres_MD = [13.66459995076317, 6.566501746216531, -2.772824757106873]
Vols_MD = [1.520281143, 1.59629509, 1.83573934]

#High Density
#pres_HD = [mean_MNS_HD, mean_DEF_HD, mean_PLS_HD, -0.4378931877775994]
pres_HD = [6.002016630400996, 3.8251471518987343, 0.9294700340225073, -0.4378931877775994]
Vols_HD = [4.988283151254265, 5.237697321417083,6.023351928537083, 6.808952248]


#Third-order Birch-Murnaghan Equation of State
def func(V, V_0, B_0, B_0_p):
    return (1.5*B_0)*((V_0/V)**(7/3) - (V_0/V)**(5/3))*(1 + (3/4)*(B_0_p - 4)*(((V_0/V)**(2/3))-1))

#Fit the curve
coefs_LD = curve_fit(func, Vols_LD, pres_LD, p0 = [2.631, 1, 1])
coefs_MD = curve_fit(func, Vols_MD, pres_MD, p0 = [2.631, 1, 1])
coefs_HD = curve_fit(func, Vols_HD, pres_HD, p0 = [2.631, 1, 1])

#Plot the curve
P_func_LD = func(np.linspace(Vols_LD[0], Vols_LD[-1], 100), coefs_LD[0][0], coefs_LD[0][1], coefs_LD[0][2])
P_func_MD = func(np.linspace(Vols_MD[0], Vols_MD[-1], 100), coefs_MD[0][0], coefs_MD[0][1], coefs_MD[0][2])
P_func_HD = func(np.linspace(Vols_HD[0], Vols_HD[-1], 100), coefs_HD[0][0], coefs_HD[0][1], coefs_HD[0][2])
```
